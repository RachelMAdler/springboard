# Capstone Project 2: Proposal

## Problem
Individuals often post sarcastic or ironic messages on social media sites, and businesses (and sometimes governments) need to be able to distinguish between what is intended literally and what is not.  The model generated for this project would enable organizations to perform this classification.

## Client
Businesses and government organizations that use Twitter data will benefit from this project.  The model generated by this project would enable such organizations to identify unlabeled sarcastic tweets and distinguish them from literal tweets.

## Data
The tweets for this project were gathered between July and October of 2017 via the Twitter API.  The most recent 3,200 tweets made by the ~ 3,000,000 followers of John Oliver were acquired (over 2 billion total tweets).  Of those tweets, approximately 30,000 contain the sarcasm hashtag, which will be used as the true label.  In addition, profile information was gathered for each user, including number of friends, number of followers, date account was created, and location.
 
## Approach
The training data will consist of tweets with #sarcam and tweets without #sarcasm (but with exactly one other hashtag).  Because there are far fewer tweets with #sarcasm compared to those without, tweets with #sarcasm will be oversampled.  Other measures will be taken to address this problem, such as adjusting class weights in the model and using a metric other than accuracy, such as AUC.  I will use the gensim and spaCy libraries to perform text normalization and phrase modeling, and to generate a word2vec model.  I will additionally perform topic modeling using LDA.

## Deliverables
The deliverables of this project will include the code for all analyses as well as a brief report of the results and a slide deck summarizing the findings.
